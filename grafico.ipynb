```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dados1 = pd.read_csv('MD - 23.1 - 317902-Exercícios de Fixação e Aprendizagem 1-notas.csv')
dados1 = dados1.drop(54)
```

```python
dados2 = pd.read_csv('MD - 23.1 - 317903-Exercícios de Fixação e Aprendizagem 1-notas.csv')
dados2 = dados2.drop(60)

```

```python
dados3 = pd.read_csv('MD - 23.1 - 322127-Exercícios de Fixação e Aprendizagem 1-notas.csv')
dados3 = dados3.drop(41)
```

```python
dadosv = pd.concat([dados1, dados2], axis=0)
dadosV = pd.concat([dadosv, dados3], axis= 0)
dadosV.replace(',', '.', regex=True)
notas = pd.to_numeric(dadosV['Avaliar/10,00'].str.replace(',', '.'), errors = 'raise')
notas
```

```python
intervalos = np.arange(0.00, 10.50, 00.50)

hist, bins = np.histogram(notas, bins = intervalos)
```

```python
plt.figure(figsize=(10, 5))

plt.bar(intervalos[:-1], hist, width=0.5, align='edge', edgecolor='black')

plt.xlabel('Intervalo de Nota')
plt.ylabel('Quantidade de Alunos')
plt.title('Distribuição de Notas dos Alunos de Matemática Discreta na EFA 1')

plt.xticks(intervalos)

for i, v in enumerate(hist):
    plt.text(intervalos[i] + 0.2, v + 0.1, str(v), color='black', fontsize=10, fontweight='bold')

plt.show()
```
